```python
import pandas as pd
```

```python
cp = pd.read_csv('../companies_data_TH.csv')
```

```python
sim = pd.read_csv('../nonnormalized_complete.csv')
```

```python
sim.groupby('company')['value'].sum()
```

```python
sim.head()
```

```python
import ast
```

```python
sim['areas'] = sim['areas'].apply(lambda x: ast.literal_eval(x))
```

```python
inf = sim.groupby('company')[['value', 'areas']].sum()
```

```python
cp = cp.join(inf, on='company_name', how='left')
cp.head()
```

```python
cp.sort_values('value', ascending=False).head(15)
```
