```python
import ast

subject_areas_df = df['subject_areas'].apply(lambda x: ast.literal_eval(x))
```

```python
areas = set()
areas_freq = dict()

for idx, row in enumerate(subject_areas_df):

    areas = areas.union(set(row))

    for i in row:
        if i not in areas_freq:
            areas_freq[i] = 1
        else:
            areas_freq[i] += 1
```
